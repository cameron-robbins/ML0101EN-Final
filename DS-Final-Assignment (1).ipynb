### Feature selection


Let's define feature sets, X:


```python
X = df1[['G', 'W', 'ADJOE', 'ADJDE', 'BARTHAG', 'EFG_O', 'EFG_D',
       'TOR', 'TORD', 'ORB', 'DRB', 'FTR', 'FTRD', '2P_O', '2P_D', '3P_O',
       '3P_D', 'ADJ_T', 'WAB', 'SEED', 'windex']]
X[0:5]
```

What are our lables? Round where the given team was eliminated or where their season ended (R68 = First Four, R64 = Round of 64, R32 = Round of 32, S16 = Sweet Sixteen, E8 = Elite Eight, F4 = Final Four, 2ND = Runner-up, Champion = Winner of the NCAA March Madness Tournament for that given year)|


```python
y = df1['POSTSEASON'].values
y[0:5]
```

## Normalize Data 


Data Standardization gives data zero mean and unit variance (technically should be done after train test split )


```python
X= preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

## Training and Validation 


Split the data into Training and Validation data.


```python
# We split the X into train and test to find the best k
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Validation set:', X_val.shape,  y_val.shape)
```

# K Nearest Neighbor(KNN)

<b>Question  1 </b> Build a KNN model using a value of k equals five, find the accuracy on the validation data (X_val and y_val)


You can use <code> accuracy_score</cdoe>


```python
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

#Build model and predict the model 
k=5 
neigh5 = KNeighborsClassifier(n_neighbors = k).fit(X_train, y_train)

#lets set y_hat5 to the predictions
y_hat5 = neigh5.predict(X_val) 

print("Train set Accuracy:", accuracy_score(y_train, neigh5.predict(X_train)))
print("Test set Accuracy: ", accuracy_score(y_val, y_hat5))
```

The output shows us that the model accuracy is 66.67% accurate on the validation set, which means:

the model predicts whether a team will make the Final Four about 2/3 of the time...

Remember from the instructions that in the sports betting industry an accuracy rate of anything over 55% is considered good.

<b>Question  2</b> Determine and print the accuracy for the first 15 values of k on the validation data:


```python
for k in range(1, 16):
    model = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    y_pred = model.predict(X_val)
    acc = accuracy_score(y_val, y_pred)
    print(f"k={k:2d}: Accuracy = {acc:.4f}")
```

# Decision Tree


The following lines of code fit a <code>DecisionTreeClassifier</code>:


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

<b>Question  3</b> Determine the minumum   value for the parameter <code>max_depth</code> that improves results 


```python
# Start with initial values
best_acc = 0     # Keep track of the best accuracy we've seen
best_depth = 1   # Keep track of which depth gave us the best accuracy

# Try depths from 1 to 15
for depth in range(1, 16):
    # Create a decision tree with current depth
    dt = DecisionTreeClassifier(max_depth=depth, random_state=4)

    #Train the model
    dt.fit(X_train,y_train)

    #Make prediction
    y_pred = dt.predict(X_val)

    #Calculate how accurate these predictions are 
    acc = accuracy_score(y_val, y_pred)

    #If the accuracy is better than our previous best 
    if acc > best_acc:
        best_acc = acc   #Save this accuracy as our new best
        best_depth = depth #Remember which depth gave us this accuracy 

        #Print our results

        print(f"Minimum max_depth with best validation accuracy: {best_depth}")
        print(f"Validation Accuracy: {best_acc:.4f}")
```

# Support Vector Machine


<b>Question  4</b> Train the support  vector machine model and determine the accuracy on the validation data for each kernel. Find the kernel (linear, poly, rbf, sigmoid) that provides the best score on the validation data and train a SVM using it.


```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python

kernels = ['linear', 'poly', 'rbf', 'sigmoid']
best_accuracy = 0
best_kernel = ''

for kernel in kernels:
    # Initialize and train the SVM model
    model = SVC(kernel=kernel)
    model.fit(X_train, y_train)
    
    # Predict on validation data
    y_pred = model.predict(X_val)
    acc = accuracy_score(y_val, y_pred)
    
    # Track the best kernel
    if acc > best_accuracy:
        best_accuracy = acc
        best_kernel = kernel
    
    # Print results for each kernel
    print(f"Kernel: {kernel} \t Validation Accuracy: {acc:.4f}")
```

The polynomial kernel was the best choice for this dataset...
It performed the best with about 67% accuracy on the validation data. This means:

- It correctly predicted whether a team would make the Final Four about 2/3 of the time
- It significantly outperformed the linear kernel
- It did somewhat better than both RBF and sigmoid kernels

# Model Evaluation using Test set


```python
from sklearn.metrics import f1_score, jaccard_score
# for f1_score please set the average parameter to 'micro'

# Calculate F1 score with micro average as specified
f1 = f1_score(y_val, y_val_pred, average='micro')
print(f"F1 Score (micro): {f1:.4f}")

# Calculate Jaccard score using the provided jaccard_index function
jaccard = jaccard_score(y_val, y_val_pred, average='micro')
print(f"Jaccard Score: {jaccard:.4f}")
```

```python
def jaccard_index(predictions, true):
    if (len(predictions) == len(true)):
        intersect = 0;
        for x,y in zip(predictions, true):
            if (x == y):
                intersect += 1
        return intersect / (len(predictions) + len(true) - intersect)
    else:
        return -1
```
